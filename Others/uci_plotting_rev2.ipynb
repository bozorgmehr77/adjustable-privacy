```python
# AUC new calculation:
from scipy.spatial import ConvexHull, convex_hull_plot_2d

def calc_auc(pts_x_in, pts_y_in, start_x, end_x, start_y, end_y):
  ref_x = end_x
  ref_y = start_y

  pts_x = pts_x_in[:]
  pts_y = pts_y_in[:]

  pts_x.insert(0,start_x)
  pts_x.append(end_x)
  x_for_real = np.array(pts_x)
  pts_x.append(ref_x)
  x = np.array(pts_x)

  pts_y.insert(0,start_y)
  pts_y.append(end_y)
  y_for_real = np.array(pts_y)
  pts_y.append(ref_y)
  y = np.array(pts_y)

  points = np.vstack((x, y)).T
  hull = ConvexHull(points)
  convex_area = hull.volume

  # plt.plot(points[:,0], points[:,1], 'o')
  # for simplex in hull.simplices:
  #   plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
  # # plt.show() 

  real_area = trapz(y_for_real, x_for_real, dx=0.001)

  #area = PolyArea2D(pts)
  return convex_area, real_area
```

```python
ac_area, ar_area = calc_auc(ae_adv, ae_utl, 0, end_x, 0, end_y)
uc_area, ur_area = calc_auc(uae_adv, uae_utl, 0, end_x, 0, end_y)
vc_area, vr_area = calc_auc(vae_adv, vae_utl, 0, end_x, 0, end_y)
bc_area, br_area = calc_auc(b_vae_adv, b_vae_utl, 0, end_x, 0, end_y)
oc_area, or_area = calc_auc(adv_acc_values, utl_acc_values, 0, end_x, 0, end_y)
print("a|convex area: " + "%.4f" % round(ac_area, 4) + "| real area: " + "%.4f" % round(ar_area, 4))
print("u|convex area: " + "%.4f" % round(uc_area, 4) + "| real area: " + "%.4f" % round(ur_area, 4))
print("v|convex area: " + "%.4f" % round(vc_area, 4) + "| real area: " + "%.4f" % round(vr_area, 4))
print("b|convex area: " + "%.4f" % round(bc_area, 4) + "| real area: " + "%.4f" % round(br_area, 4))
print("o|convex area: " + "%.4f" % round(oc_area, 4) + "| real area: " + "%.4f" % round(or_area, 4))
```
